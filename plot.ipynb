```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_csv("toSubmit2/output.txt", header=None, names=['x','y','U','V'], sep=r"\s+")

print(df['U'])
```

```python
Nx, Ny = 101, 101
df['U'].values.reshape(Nx, Ny).T
```

```python
fig, axs = plt.subplots(1,2)
ax = axs[0]
c = ax.pcolor(df['U'].values.reshape(Nx, Ny).T)
ax.set_title('U value')
fig.colorbar(c, ax=ax)

ax = axs[1]
c = ax.pcolor(df['V'].values.reshape(Nx, Ny).T)
ax.set_title('V value')
fig.colorbar(c, ax=ax)
fig.tight_layout()
fig.suptitle(f"Iterating for Nx={Nx}, Ny={Ny}\n")
plt.savefig(f"test1.png")
```
